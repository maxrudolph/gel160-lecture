```python
import numpy as np
from numpy.fft import rfft
import matplotlib.pyplot as plt
```

```python
# make a sawtooth wave
# Define the period T
T = 1./440.
# define the duration of the signal, for plotting
tplot = 4*T
tvec = np.linspace(-tplot,tplot,1000)
dt = tvec[1]-tvec[0]

h = 1/T # amplitude of the signal - this will producee h*T = 1
def sawtooth(t,h=h):
    t = (t-T) % (2*T) - T
    return h*t
plt.figure()
plt.plot(tvec,sawtooth(tvec))
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
signal = sawtooth(tvec)
#signal = np.sin(np.pi*2.*tvec/T)
n=len(signal)
Y = np.fft.fft(signal,n)
f = np.fft.fftfreq(n,dt)   # this gets the frequencies corresponding to each component
```

```python
plt.figure()

pmask = f>=0
plt.plot(f[pmask],np.abs(Y[pmask]),'.')
plt.ylabel('Amplitude')
plt.xlabel('Frequency')
#plt.plot(f,Y.imag,'.')
plt.show()
```
