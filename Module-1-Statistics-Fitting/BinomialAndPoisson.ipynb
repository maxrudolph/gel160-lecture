```python
import numpy as np
from scipy.special import comb
import matplotlib.pyplot as plt
```

```python
# Calculate and plot the binomial distribution

p=1.0/6.0 # probability of a single event occuring
N=100  # Number of observations

xmax = int(np.ceil(np.min([N+1,4*N*p])))

x = np.arange(0,xmax)
pb = np.zeros(x.shape)
for i in x:
    q = 1-p
    pb[i] = p**i * q**(N-i) * comb(N,i)
# mean and sigma2    
mu = N*p
sigma2 = N*p*(1-p)
print('St. Dev=',np.sqrt(sigma2))

plt.figure()
plt.plot(x,pb,label='Prob. x events')
ymax = pb.max()
plt.plot([mu,mu],[0,ymax],'k--',label='mean value')
plt.legend()
plt.xlabel('x')
plt.ylabel('prob(x)')
plt.show()
```
