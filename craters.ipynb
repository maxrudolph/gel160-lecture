```python
import numpy as np
import matplotlib.pyplot as plt
```

The database of lunar impact crater properties was obtained from https://www.lpi.usra.edu/scientific-databases/
and is a compilation of literature values. The data plotted are from columns 8  and 14 of the Excel spreadsheet

```python
crater_file = 'Crater_Database.csv'
diameter, depth = np.loadtxt(crater_file,delimiter=',',unpack=True,skiprows=1)
# Select only the small craters
#idx = diameter>=10.
#diameter = diameter[idx]
#depth = depth[idx]
```

```python
plt.figure()
plt.plot(diameter,depth,'.')
plt.xlabel('Diameter (km)')
plt.ylabel('Depth (km)')
plt.show()
```

```python
plt.figure()
plt.plot(diameter,depth,'.')
plt.xlabel('Diameter (km)')
plt.xscale('log')
plt.yscale('log')
plt.ylabel('Depth (km)')
plt.show()
```

```python
ylog = np.log(diameter)
xlog = np.log(depth)

fc = np.polyfit(xlog,ylog,1)
```

```python
fc
```

```python
plt.figure()
plt.plot(xlog,ylog,'.')
plt.plot(xlog,(xlog*fc[0]+fc[1]),'.')
plt.xlabel('Diameter (km)')
#plt.xscale('log')
#plt.yscale('log')
plt.ylabel('Depth (km)')
plt.show()
```
