```python
import os
from skimage import io
import skimage
filename = os.path.join(skimage.data_dir, 'camera.png')
camera = io.imread(filename)

import matplotlib.pyplot as plt
plt.figure()
plt.imshow(camera,cmap='gray')
plt.colorbar()
plt.show()
```
