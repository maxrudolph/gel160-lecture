```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eqfile = 'query.csv'
eqyear = []
file = open(eqfile)
ind=0;
for line in file:    
    if(ind==0):
        ind+=1
    else:
        fields = line.split(',')
        eqyear.append(float(fields[0].split('-')[0]))
eqyear = np.array(eqyear)
        
file.close()
```

```python
plt.figure()

years=np.arange(1970,2010,1)
neq = []
for year in years:
    neq.append(np.sum(eqyear == year))
neq = np.array(neq)

plt.plot(years,neq)
```
